####Due to significant correlations between variables, such as the high correlation of 78% between flavanoids and dilution, manually removing variables from the dataset can be a lengthy and tedious process. To simplify this task, we can employ Principal Component Analysis (PCA) for dimensionality reduction. PCA automatically identifies and combines variables that contribute the most to the dataset's variance, thereby eliminating the need for manual selection and streamlining the analysis.

##Dimensionality Reduction with PCA

```python
# Converting data to numpy array
wine_ary=wine_data.values
wine_ary
```

```python
# Normalizing the numerical data
wine_norm=scale(wine_ary)
wine_norm
```

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color='magenta')
```

```python
# Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(12,8))
sns.scatterplot(data=final_df)
```

#K-Means Clustering

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range(1,6):
  kmeans=KMeans(n_clusters=i,random_state=2)
  kmeans.fit_transform(wine_norm)
  wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

##Build Cluster algorithm using K=3

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=0).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign CLuster to the dataset
wine_data2=wine.copy()
wine_data2['clusters3id']=clusters3.labels_
wine_data2
```
