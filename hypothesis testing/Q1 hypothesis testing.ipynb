```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import stats
import statsmodels.api as sm
import seaborn as sns
import warnings
warnings.filterwarnings('ignore') 
```

```python
# null hypothesis(H0)      = mew1=mew2
# Alternative Hypthesis (H1) = mew1 != mew2
```

```python
cutlets=pd.read_csv('Cutlets.csv')
```

```python
unitA=pd.Series(cutlets.iloc[:,0])
unitA
```

```python
unitB=pd.Series(cutlets.iloc[:,1])
unitB
```

```python
#ploting data
plt.figure(figsize=(9,6))
plt.subplot(121)
plt.boxplot(cutlets.iloc[:,0])
plt.title('unitA')
plt.figure(122)
plt.boxplot(cutlets.iloc[:,1])
plt.title('unitB')
plt.show()
```

```python
## ploting Histogram 
plt.subplots(figsize=(9,6))
plt.subplot(121)
plt.hist(cutlets.iloc[:,0],bins=15)
plt.title('unit A')
plt.subplot(122)
plt.hist(cutlets.iloc[:,1],bins=15)
plt.title('unit B')
plt.show()
```

```python
plt.figure(figsize=(8,6))
label=('unit A','unit B')
sns.distplot(cutlets.iloc[:,0],kde= True)
sns.distplot(cutlets.iloc[:,1],hist=True)
plt.legend(label)
```

```python
# for noraml distribution we will use qq plot to see whether it is following normal distribution or not
sm.qqplot(cutlets['Unit A'],line='q')
plt.title('Unit A')
sm.qqplot(cutlets['Unit B'], line='q')
plt.title('Unit B')
plt.show()
```

```python

# Two sample so Two Tail Test
p_value=stats.ttest_ind(cutlets['Unit A'],cutlets['Unit B'])
p_value=p_value[1]
print('p_value' '=', p_value)
```
